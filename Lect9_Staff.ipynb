# Exercise-1: Checkerboard Pattern
Plear write a python code to output a matrix with the following style:

1699929822024.png

```python
import numpy as np
import matplotlib.pyplot as plt

def generate_checkerboard(size):
    pattern = np.zeros((size, size))
    for i in range(size):
        for j in range(size):
            if (i % 2 == 0 and j % 2 == 0) or (i % 2 == 1 and j % 2 == 1):
                pattern[i, j] = 1
    return pattern

# specify the size of the checkerboard
size = 8

checkerboard = generate_checkerboard(size)

plt.imshow(checkerboard, cmap='gray', interpolation='nearest')
plt.show()
```
